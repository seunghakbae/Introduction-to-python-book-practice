# 파이썬 표준 라이브러리

파이썬의 두드러진 점 중 하나는 배터리 포함이라는 철학으로 유용한 작업을 처리하는 많은 표준 라이브러리 모듈이 있는 것이다. 

### 누락된 키 처리하기 : setdefault(), defaultdict()

존재하지 않는 키로 딕셔너리에 접근하면 예외가 발생한다. 기본값을 반환하는 딕셔너리의 get()함수를 사용하면 이 예외를 피할 수 있다. setdefault() 함수는 get()함수와 같지만, 키가 누락된 경우 딕셔너리에 항목을 할당할 수 있다. 

```python
periodic_table = {'Hydrogen':1, 'Helium':2}
```

```python
print(periodic_table)
```

```python
# get함수를 사용하는 방법. 
periodic_table.get('Hydrogen')
```

```python
# get함수를 사용하되 존재하지 않는 키를 입력하는 경우. 
periodic_table.get('Carbon')
```

```python
# setdefault() 사용하는 경우
carbon = periodic_table.setdefault('Carbon', 12)
carbon
print(carbon)
print(periodic_table) # 존재하지 않는 경우, 딕셔너리에 항목을 할당할 수 있다. 
```

```python
# 현재 존재하고 있는 경우, 바뀌지 않고 기존의 값을 반환한다. 
helium = periodic_table.setdefault('Helium',947)
helium
```

```python
periodic_table
```

### defaultdict() 함수

defaultdict() 함수도 비슷하다. 다른 점은 딕셔너리를 생성할 때, 모든 새 키에 대한 기본값을 먼저 지정한다는 것이다. 이 함수의 인자는 함수다. 다음 예제에서 int()함수를 호출하고 정수 0을 반환하는 함수 int를 전달해보자. 

```python
from collections import defaultdict
periodic_table = defaultdict(int)
```

```python
periodic_table['Hydrogen'] = 1
periodic_table['Lead'] # 아무것도 지정되지 않으면 0이 저장된다. 
```

```python
periodic_table
```

```python
from collections import defaultdict

def no_idea():
    return 'Huh?'
```

```python
bestiary = defaultdict(no_idea)
```

```python
bestiary['A'] = 'Abominable'
bestiary['B'] = 'Basilisk'
```

```python
bestiary['C']
```

```python
# 키 값만 주면 저절로 value값이 입력된다. 
bestiary
```

```python
# lambda를 사용하여 인자에 기본값을 만드는 함수를 정의할 수 있다. 
bestiary = defaultdict(lambda : 'Huh?')
```

```python
bestiary['A']
```

```python
bestiary
```

```python
# 카운터를 만들기 위해 아래와 같이 int함수를 사용할 수 있다. 
food_counter = defaultdict(int)
for food in ['spam', 'spam', 'eggs', 'spam']:
    food_counter[food] += 1
```

```python
food_counter
```

```python
for food, count in food_counter.items():
    print(food, count)
```

```python
# 만약 food_counter에 default_dict이 아니라 그냥 dictionary라면 파이썬은 예외를 발생시킨다. 딕셔너리가 초기화되지 않았기 때문이다. 
```

```python
dict_counter = {}

for food in ['spam', 'spam', 'eggs', 'spam']:
    if not food in dict_counter:
        dict_counter[food] = 0
    dict_counter[food] += 1
```

```python
dict_counter
```

### 항목 세기: Counter()

표준라이브러리에는 항목을 셀 수 있는 함수가 여러 개 있다. 

```python
from collections import Counter
breakfast = ['spam', 'spam', 'eggs', 'spam']
```

```python
breakfast_counter = Counter(breakfast)
```

```python
breakfast_counter
```

```python
# most_common()함수는 모든 요소를 내림차순으로 반환한다. 
breakfast_counter.most_common()
```

```python
breakfast_counter.most_common(1)
```

```python
# Counter는 결합할 수 있다. 
lunch = ['eggs', 'eggs', 'bacon']
lunch_counter = Counter(lunch)
```

```python
# + 연산자를 사용하면 결합이 가능하다. 
lunch_counter + breakfast_counter
```

```python
# - 연산자를 사용하면 뺼 수 있다. 
breakfast_counter - lunch_counter
```

```python
# set과 마찬가지로 인터섹션 연산자 &를 사용하면 공통된 항목을 얻을 수 있다. 
breakfast_counter & lunch_counter
```

```python
# Union연산자 | 를 사용하면 모든 항목을 얻을 수 있다. 
breakfast_counter | lunch_counter
```

### 키 정렬하기 : OrderedDict()

이 책의 앞부분에서 딕셔너리의 키 순서는 예측할 수 없는 것을 봤다. 

```python
quotes = {'Moe':'A wise guy, huh?', 'Larry' : 'Ow!', 'Curly' : "Nyuk Nyuk!"}
```

```python
for quote in quotes:
    print(quote)
```

OrderedDict() 함수는 키의 추가 순서를 기억하고, 이터레이터로부터 순서대로 키값을 반환한다. (키, 값) 튜플의 시퀀스로부터 OrderedDict을 생성해보자.

```python
from collections import OrderedDict
```

```python
quotes = OrderedDict([('Moe','A wise guy, huh?'), ('Larry','Ow!'), ('Curly',"Nyuk Nyuk!")])
```

```python
for stooge in quotes:
    print(stooge)
```

### deque

데크는 스택과 큐의 기능을 모두 가진 출입구가 양 끝에 있는 큐다. 데크는 시퀀스의 양 끝으로부터 항목을 추가하거나 삭제할 때 유용하게 쓰인다. 

```python
def palindrome(word):
    from collections import deque
    dq = deque(word)
    while len(dq) > 1:
        if dq.popleft() != dq.pop():
            return False
    return True
```

```python
palindrome('a')
```

```python
palindrome('abba')
```

```python
palindrome('close')
```

```python
# 이렇게도 체크를 할 수 있다. 
def palindrome(word):
    return word == word[::-1]
```

```python
palindrome('a')
```

```python
palindrome('abba')
```

```python
palindrome('close')
```

### 코드 구조 순회하기 : itertools
itertools는 특수 목적의 이터레이터 함수를 포함하고 있다. for ... in루프에서 이터레이터 함수를 호출할 때 함수는 한 번에 한 항목을 반환하고, 호출 상태를 기억한다. 

```python
import itertools
```

```python
for item in itertools.chain([1,2], ['a', 'b']):
    print(item)
```

```python
# cycle() 함수는 인자를 순환하는 무한 이터레이터이다. 
for item in itertools.cycle([1,2]):
    print(item)
```

```python
# accumulate() 함수는 축적된 값을 계산한다. 기본적으로 합계를 계산한다. 
for item in itertools.accumulate([1,2,3,4]):
    print(item)
```

```python
# accumulate() 함수의 두 번째 인자로 함수를 전달하여, 합계를 구하는 대신 이 함수를 사용할 수 있다. 이 함수는 두 개의 인자를 취하여
# 하나의 결과를 반환한다. 
```

```python
import itertools
```

```python
def multiply(a,b):
    return a * b
```

```python
for item in itertools.accumulate([1,2,3,4], multiply):
    print(item)
```
